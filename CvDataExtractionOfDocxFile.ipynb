```python
import re
import os
import docx
import docx2txt
# Function to extract the data from DOCX
def get_text_from_word(file_path):
    text = docx2txt.process(file_path)
    return text

def get_the_list_docx(dir_path):
    list_of_all_file = os.listdir(dir_path)
    list_of_docx = []

    for file in list_of_all_file:
        if file.endswith(".docx") and not (file.startswith("~")):
            list_of_docx.append(file)
    
    return list_of_docx
'''
    Main Function
'''
# Get list of DOCX files present in the directory
list_of_docx = get_the_list_docx("files/")

# Regex For Extracting Data
phn_ = r'[\+]?[(]?[0-9]{3}[)]?[-\s\.]?[0-9]{3}[-\s\.]?[0-9]{4,6}'
email_=r"[A-Za-z0-9]+[\._]?[a-z0-9]+[@]\w+[-]?\w+[.]\w{2,3}"

# Traverse thorugh all the documents present in the directory
for docx in list_of_docx:
    doc_string = get_text_from_word("files/" + docx)
    
    #Get all the linkedin Website in list
    result = re.finditer(r"((http|https):\/\/)+(www.)?(linkedin.com\/)+([a-zA-z]{2}\/)?[a-zA-Z0-9-]{5,30}", doc_string)
    res = " ".join(doc_string.split('\n'))
    for match_obj in result:
        print(match_obj.group())
    
    Phone=re.findall(phn_,doc_string)
    Email=re.findall(email_,doc_string)
    print(Phone)
    print(Email)
    print("\n\n\n\n\n\n**********************************************\n\n\n\n\n")
    
```

```python
import docx2txt
import re


SKILLS = [
    'machine learning',
    'data science',
    'python',
    'Creative design',
    'word',
    'excel',
    'Strong decision maker',
    'English',
    'web developer',
    'Innovative',
    'html',
    'Service-focused',
    'Complex problem solver',
    'css',
    'Project management'
    
]


def get_text_from_doc(file_path):
    file_content=docx2txt.process(file_path)
    " ".join(file_content.split("\n"))
    return file_content

def find_skills(text):
    skills = []
    for i in skills:
        pattern = re.compile(i)
        if bool(re.search(pattern, text.lower())):
            skills.append(i)
    return skills
        

text = get_text_from_doc("Resume-3.docx")
skills = find_skills(text)
print(SKILLS)

```
